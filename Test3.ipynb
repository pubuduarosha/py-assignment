```python
# Exclude rows with zero square footage
data = data[data['sq__ft'] > 0]
```

```python
data.shape
```

```python
# Calculate the average price per square foot
data['price_per_foot'] = data['price'] / data['sq__ft']
average_price_per_foot = data['price_per_foot'].mean()
```

```python
average_price_per_foot
```

```python
# Filter properties sold for less than the average price per square foot
filtered_data = data[data['price_per_foot'] < average_price_per_foot]
```

```python
filtered_data
```

```python
# Save the filtered data to a new CSV file
filtered_data.to_csv('filtered_properties.csv', index=False) 
```
